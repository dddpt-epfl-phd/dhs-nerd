```python
from csv import DictReader
import sys
sys.path.append("../../scripts")

from file_paths import S2_CLEF_HIPE_PRED_FILE, S2_CLEF_HIPE_TRUE_FILE, localize
language = "fr"
```

```python
def clef_hipe_load_conllu_tsv_to_dict(filepath):
    with open(filepath) as csvfile:
        conllu_tsv = [r for r in DictReader(csvfile, delimiter="\t")]
    conllu_tsv = [r for r in conllu_tsv if len(r["TOKEN"])!=0 and not r["TOKEN"].startswith("#")]
    return conllu_tsv
```

```python

pred_conllu = clef_hipe_load_conllu_tsv_to_dict(localize(S2_CLEF_HIPE_PRED_FILE, language))
true_conllu = clef_hipe_load_conllu_tsv_to_dict(localize(S2_CLEF_HIPE_TRUE_FILE, language))
```
